```python
results["date"] = pd.to_datetime(results["date"])
```

```python
results.info()
```

```python
def get_result(row):
    if row["home_score"] > row["away_score"]:
        return "Home Win"
    elif row["home_score"] < row["away_score"]:
        return "Away Win"
    else:
        return "Draw"
```

```python
results["result"] = results.apply(get_result, axis=1)
```

```python
results[["home_team", "away_team", "home_score", "away_score", "result"]].head()
```

```python
result_map = {
    "Home Win": 1,
    "Draw": 0,
    "Away Win": -1
}

results["result_label"] = results["result"].map(result_map)
```
